```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, levenshtein

#Dans ce code, nous avons utilisé la fonction subtract() pour obtenir les utilisateurs de la table "Users_" qui ne se trouvent
#pas dans la liste des utilisateurs ayant les mêmes adresses que les rendez-vous. Ainsi, nous obtenons à la fois les utilisateurs
#avec les mêmes adresses et ceux avec des adresses différentes.

# Créer une session Spark
spark = SparkSession.builder \
    .appName("Agenda App") \
    .getOrCreate()

# Rejoindre les tables sur les colonnes adresse et code_postal
joined_df = df.join(df1, 
                                     (df["Jobs_address"] ==df1["users_address"]) &
                                     (df["Jobs_postcode"] == df1["users_postcode"]),
                                     "inner")

# Calculer la distance de Levenshtein entre les adresses
joined_df = joined_df.withColumn("address_similarity", levenshtein(col("Jobs_address"), col("users_address")))

# Filtrer les résultats pour une distance minimale acceptable
min_similarity_threshold = 3  # Vous pouvez ajuster ce seuil selon votre besoin
matching_users_df = joined_df.filter(col("address_similarity") <= min_similarity_threshold)

# Obtenir la liste des utilisateurs avec les mêmes adresses
matching_users_list = matching_users_df.select("users_full_name", "users_email_address", "users_address", "users_city", "users_postcode")

# Obtenir la liste complète des utilisateurs
all_users_list = df1.select("users_full_name", "users_email_address", "users_address", "users_city", "users_postcode")

# Trouver les utilisateurs avec des adresses différentes
non_matching_users_df = all_users_list.subtract(matching_users_list)

# Afficher ou enregistrer les résultats
non_matching_users_df.show()

display(non_matching_users_df)
```
